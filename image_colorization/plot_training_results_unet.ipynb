# Plot the results of the training

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# U-Net training results

## Select files

```python
training_log_path = './results/test/training_losses_unet_20250121_174041.csv'
training_df = pd.read_csv(training_log_path)
```

## Plot losses

```python
# Create subplots with a 2x1 grid for the combined plot and individual plots
fig, ax1 = plt.subplots(1, 1, figsize=(18, 6))

# Plot both losses in one plot
ax1.plot(training_df['Epoch'], training_df['Train Loss'], label='Train Loss', color='orange')
ax1.plot(training_df['Epoch'], training_df['Validation Loss'], label='Validation Loss', color='blue')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.set_title('Train and Validation Losses for U-Net')
ax1.legend()
ax1.grid(True)

# Create a new figure for the individual plots
fig2, (ax2, ax3) = plt.subplots(1, 2, figsize=(18, 6))

# Plot Train Loss
ax2.plot(training_df['Epoch'], training_df['Train Loss'], label='Train Loss', color='orange')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.set_title('Train Loss')
ax2.legend()
ax2.grid(True)

# Plot Validation Loss
ax3.plot(training_df['Epoch'], training_df['Validation Loss'], label='Validation Loss', color='blue')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Loss')
ax3.set_title('Validation Loss')
ax3.legend()
ax3.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```
