# Plot the results of the training

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# U-Net training results

## Select files

```python
training_log_path = '../results/20250206_102039/training_losses_unet_20250206_102039.csv'
training_df = pd.read_csv(training_log_path)
```

## Plot losses

```python
def plot_losses(df):
    # Plot losses
    fig, axes = plt.subplots(4, 1, figsize=(18, 24))

    metrics = ['MSE', 'MAE', 'SSIM', 'PSNR']
    colors = ['orange', 'blue']

    for i, metric in enumerate(metrics):
        ax = axes[i]
        ax.plot(df['Epoch'], df[f'Train Loss ({metric})'], label=f'Train Loss ({metric})', color=colors[0])
        ax.plot(df['Epoch'], df[f'Validation Loss ({metric})'], label=f'Validation Loss ({metric})', color=colors[1])
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(f'Train and Validation Losses for {metric}')
        ax.legend()
        ax.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
plot_losses(training_df, 'U-Net training loss')
```

```python
def plot_losses_2x2(training_df, title):
    fig, axes = plt.subplots(2, 2, figsize=(18, 12))
    fig.suptitle(title, fontsize=16)

    metrics = ['MSE', 'MAE', 'SSIM', 'PSNR']
    colors = ['orange', 'blue']
    positions = [(0, 0), (0, 1), (1, 0), (1, 1)]

    for i, metric in enumerate(metrics):
        row, col = positions[i]
        ax = axes[row, col]
        ax.plot(training_df['Epoch'], training_df[f'Train Loss ({metric})'], label=f'Train Loss ({metric})', color=colors[0])
        ax.plot(training_df['Epoch'], training_df[f'Validation Loss ({metric})'], label=f'Validation Loss ({metric})', color=colors[1])
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(f'Train and Validation Losses for {metric}')
        ax.legend()
        ax.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# Usage
plot_losses_2x2(training_df, 'U-Net training loss')
```

# Real results


```python
training_log_path_ssim = '../results/training_losses_unet_20250208_093148.csv'
training_ssim_df = pd.read_csv(training_log_path_ssim)
plot_losses_2x2(training_ssim_df, 'SSIM training loss, 100 Epochs, 20250208_093148')
```

```python
training_log_path_psnr = '../results/training_losses_unet_20250212_102524.csv'
training_psnr_df = pd.read_csv(training_log_path_psnr)
plot_losses_2x2(training_psnr_df, 'PSNR training loss, 100 Epochs, 20250208_093148')
```
